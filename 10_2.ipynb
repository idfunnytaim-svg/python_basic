```python
import random
balls=list(range(1,46))
random.choice(balls)
```

```python
for i in range(0,1000000):
    temp=balls[0]
    r_num=int(random.random()*len(balls))
    balls[0]=balls[r_num]
    balls[r_num]=temp
print(balls)
print(balls[:6])
```

```python
#random동안 sample()함수를 사용
print(random.sample(range(1,46), 6))
```

```python
for i in range(10):
    lotto=random.sample(range(1,46),6)
    lotto.sort()
    print(f"{i}번째 로또번호는 {lotto}")
```
