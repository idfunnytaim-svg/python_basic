```python
# 64bit=8byte
# 한글같은 경우 한글자당1byte
# 정수 범위 -128~0~127 =>256개의 int
# 파이썬은 정수크기가 정해지지않는다. 필요에 따라서 계속 커질수 있다.
# 파이썬은 숫자등등을 객체로 저장한다.
# 127에서 +1을 하면 -128 오버플로우가 된다.
```

```python
n = 10
print(n.bit_length()) # 10의 이진수 값(1010)의 길이

# bin()함수는 인자로 입력된 10진수의 2진수 문자열을 반환한다.
# 이진수 문자열 앞에 binary를 뜻하는 "0b"접두사를 앞에 붙여서 반환한다.
print(bin(n)) #0b1010 "0b" + "1010"

# bit_length()메서드는 n이라는 숫자 객체를 만들어야 존재하는 것이고,
# bin() 함수는 n객체와 상관없는 독립된 함수 객체이다.

print(bin(9)) # n과 상관없이 독립적으로 사용가능
```
