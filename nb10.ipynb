```python
# 1급 함수(First-Class Function)​란 함수를
# 일반 데이터(객체)처럼 다룰 수 있는 함수를 말한다.
# 파이썬에서는 함수도 객체이므로 1급 함수를 지원한다.
# 즉, 함수를 다음과 같이 사용할 수 있다.
# 1) 변수에 저장할 수 있다. 
# 2) 함수의 인수(argument)로 전달할 수 있다. 
# 3) 함수의 반환값(return)으로 사용할 수 있다. 
# 4) 자료구조(list, dict 등)에 저장할 수 있다.
```

```python
def hello():
    print("Hello")

f = hello      # 함수 자체를 저장 (괄호 없음)

f()
```

```python
def square(x):
    return x * x

def test(fun):
    return fun(5)

print(test(square))
```

```python
def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()
```

```python
def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```

```python
# 파이썬의 enumerate() 함수는 리스트, 튜플, 문자열과 같은 
# 순회 가능한(iterable) 자료형을 입력받아, 인덱스(index)와 
# 그에 해당하는 값(value)을 동시에 반환하는 매우 유용한 내장 함수입니다.
# 주로 for 반복문에서 요소의 값과 해당 요소가 
# 몇 번째에 있는지(인덱스)를 함께 알아야 할 때 사용합니다.
```

```python
names = ["홍길동", "이순신", "강감찬"]
for i, name in enumerate(names):
    print(i, name)
```

```python
names = ["홍길동", "이순신", "강감찬"]
for i, name in enumerate(names):
    print(f"({i+1})",name)
```

```python
nums = [1, 2, 3]
print(list(reversed(nums)))
nums = [0, 0, 3]
print(any(nums))#하나라도 트루인지검사
nums = [1, 2, 3]
print(all(nums))#모두트루인지검사
```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
result = zip(names, scores)
print(list(result)) # [('홍길동', 90), ('이순신', 80), ('김유신', 100)]
```

```python
names = ["홍길동", "이순신", "김유신",'ㅁ']
scores = [90, 80, 100, 120]
grades = ["A", "B", "A",'ㅠ']
print(list(zip(names, scores, grades)))
#묶을때 묶는것끼리 하나라도 부족하면 부족한부분 생략된다
```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
for name, score in zip(names, scores):
    print(name, score)
# 홍길동 90
# 이순신 80
# 김유신 100
```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = dict(zip(names, scores))

print(result) # {'홍길동': 90, '이순신': 80, '김유신': 100}
```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = {name: score for name, score in zip(names, scores)}
#딕셔너리 컴프리핸션
print(result)
```

```python
result = {} # 빈 딕셔너리 생성
for name, score in zip(names, scores):
    result[name] = score # 하나씩 키와 값으로 추가
print(result)
```
