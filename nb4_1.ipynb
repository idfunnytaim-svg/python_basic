```python
print("123".isdigit())      # 문자열이 숫자로만 구성되어 있는가?
print("Python".isalpha())   # 문자열이 영문자(또는 한글 등 문자)로만 구성되어 있는가?
print("Python123".isalnum()) # 문자열이 문자와 숫자로만 구성되어 있는가?
```

```python
name = "Python"

print(f"{name:*<15}")  # 왼쪽 정렬, 빈칸은 *로 채움
print(f"{name:*^15}")  # 가운데 정렬, 빈칸은 *로 채움
print(f"{name:*>15}")  # 오른쪽 정렬, 빈칸은 *로 채움
# 변수 사용 가능
w = 15
print(f"{name:*>{w}}")  # 오른쪽 정렬, 빈칸은 *로 채움
```
