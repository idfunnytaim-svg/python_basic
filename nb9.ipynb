```python
def func(x):
    x += 10
    print("함수 안:", x, id(x))

a = 10

func(a)

print("함수 밖:", a, id(a))

# 함수 안: 20 140714686129672
# 함수 밖: 10 140714686129352
# 함수 안의 x만 변경, 함수 밖의 a는 그대로
# # id()값이 다르다
# id(obj)는 객체의 고유 식별자(identity) 를 반환한다.
# CPython에서는 이 값이 일반적으로 
# 객체의 메모리 주소와 같다.
# id()가 같으면 같은 객체를 참조하고 있다.
# id()가 다르면 서로 다른 객체이다.
```

```python
#두 시퀀스의 교집합을 구하는 함수

def my_intersect(a, b):
    result = []
    for e in a:
        if e in b:
            result.append(e)
    return result

print(my_intersect([1, 2, 3], [2, 3, 5]))
print(my_intersect((1, 2, 3), (2, 3, 5)))
print(my_intersect({1, 2, 3}, {2, 3, 5}))
print(my_intersect({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 공통 요소 개수

def my_count_common(a, b):
    count = 0
    for e in a:
        if e in b:
            count += 1
    return count


print(my_count_common([1, 2, 3], [2, 3, 5]))
print(my_count_common((1, 2, 3), (2, 3, 5)))
print(my_count_common({1, 2, 3}, {2, 3, 5}))
print(my_count_common({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 차집합
# a에는 있고 b에는 없는 요소

def my_difference(a, b):
    result = []
    for e in a:
        if e not in b:
            result.append(e)
    return result


print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 합집합

def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)

    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
```

```python
# 포함여부확인
# value가 시퀀스 안에 있는지 확인

def my_contains(data, value):

    for e in data:
        if e == value:
            return True

    return False


print(my_contains([1, 2, 3], 2))
print(my_contains(("a", "b", "c"), "d"))
print(my_contains({"id":"abc", "pw":"1234"}, "id"))
```

```python
# 특정 값이 몇 번 나오는지 계산

def my_count(data, target):

    count = 0

    for e in data:
        if e == target:
            count += 1

    return count


print(my_count([1, 2, 1, 3, 1], 1))
print(my_count(("a", "b", "a"), "a"))
print(my_count("banana", "a"))
```

```python
# 중복 제거

def my_unique(data):

    result = []

    for e in data:
        if e not in result:
            result.append(e)

    return result


print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```

```python
# 1. 람다함수 문법
# 함수 정의는 함수선언문 이다 - 함수를 정의하는 문
# 람다는 함수 표현식 이다
# - 람다 표현식의 결과로 함수를 반환한다.

# ※ 파이썬에는
# 조건문 / 조건표현식(삼항연산자) 
# 반복문 / 반복표현식(컴프리핸션)
# 함수(문) / 람다(함수표현식)
# 가 있다.
# 람다 자체는 함수(객체)를 반환하고 람다를 호출하면 람다의 반환값이 반환된다.
#lambda 매개변수: 반환값
```

```python
# 일반 함수 정의
def add1(a, b):
    return a + b

# 람다 함수
add2 = lambda a, b : a + b # add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)

r2 = add2(1, 3)
print(r2)
```

```python
s = [('홍길동', 90), ('이순신', 85), ('강감찬', 100)]
r = sorted(s, key=lambda x: x[1], reverse=True)
# lambda x: x[1] 
# x(리스트의 요소 튜플)를 입력받아 x[1](튜플의 인덱스1값)를 반환한다.
print(r)
# [('강감찬', 100), ('홍길동', 90), ('이순신', 85)] 점수의 내림차순으로 정렬됨

# ① 원본 리스트를 하나씩 꺼낸다.
# ('홍길동', 90)
# ('이순신', 85)
# ('강감찬', 100)

# ② key 함수의 인수로 하나씩(리스트 사이즈 만큼 key함수를 반복 호출하면서 하나씩 전달)  전달한다.

# ③ sorted 내장함수는 인수로 입력받은 key함수의  반환값(x[1] 요소)으로 정렬한다.
```

```python
print(sorted(s))
```

```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True)

print(r)
```

```python
# 1) map() - 모든 요소를 변환
nums = [1, 2, 3, 4]

# 리스트의 각 요소(x)에 2를 곱한 결과를 반환
result = list(map(lambda x: x * 2, nums))

print(result) #[2, 4, 6, 8]
```

```python
# 2) filter() - 조건에 맞는 요소만 선택
nums = [1, 2, 3, 4, 5]

# 짝수인 요소만 선택
result = list(filter(lambda x: x % 2 == 0, nums))

print(result)
```

```python
# 3) sorted() - 정렬 기준 지정
names = ["banana", "apple", "kiwi"]

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key=lambda x: len(x))

print(result)
```

```python
# * stored()에 람다를 사용하면 데이터 과학에서 자주 사용되는 [dict, dict, ...] 자료구조를 원하는 키값으로 정렬이 가능하다.
students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["나이"])
print(result)

# x : 리스트의 각 딕셔너리
# x["나이"] : 각 딕셔너리의 "나이" 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.
```

```python
# "점수"를 기준으로 내림차순 정렬
result = sorted(students, key=lambda x: x["점수"], reverse=True)
print(result)
```

```python
# "이름"을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["이름"])
print(result)
```

```python
#- 여러 키를 기준으로 정렬도 가능하다
students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"]))
print(result)
```

```python
#4) min() - 기준에 따라 최소값 찾기
names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

print(result) # kiwi
```

```python
min(names)
```

```python
#5) max() - 기준에 따라 최대값 찾기
names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 긴 요소 반환
result = max(names, key=lambda x: len(x))

print(result) # banana
```

```python
max(names)
```

```python
# 6) any() - 하나라도 조건을 만족하는지 확인
nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환
result = any(map(lambda x: x % 2 == 0, nums))

print(result) # True
```

```python
# 7) all() - 모두 조건을 만족하는지 확인
nums = [2, 4, 6, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 모두 True이면 True 반환
result = all(map(lambda x: x % 2 == 0, nums))

print(result) #True
```
