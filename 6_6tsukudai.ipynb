```python
# 문제 11
# 가장 구매금액이 높은 고객 TOP 3를 출력하시오.
# 11-2) 연평균 구매금액(소수점1자리ㅠ반올림)이 
# 높은 구매고객 TOP3?
```

```python
top1_price = 0
top2_price = 0
top3_price = 0

top1_name = ""
top2_name = ""
top3_name = ""

for o in orders:
    v = o["구매금액"]
    name = o["고객명"]

    # 새로운 1등인 경우
    if v > top1_price:
        top3_price = top2_price
        top3_name = top2_name

        top2_price = top1_price
        top2_name = top1_name

        top1_price = v
        top1_name = name

    # 새로운 2등인 경우
    elif v > top2_price:
        top3_price = top2_price
        top3_name = top2_name

        top2_price = v
        top2_name = name

    # 새로운 3등인 경우
    elif v > top3_price:
        top3_price = v
        top3_name = name

print("TOP1 :", top1_name, top1_price)
print("TOP2 :", top2_name, top2_price)
print("TOP3 :", top3_name, top3_price)
```

```python
# 버블 정렬(Bubble Sort)의 오름차순(Ascending Order)이란 
# 작은 값부터 큰 값 순서대로 정렬하는 것입니다.
# 예를 들어[5, 2, 8, 1, 4]를 오름차순으로 정렬하면
# [1, 2, 4, 5, 8]이 됩니다.
# 버블 정렬의 원리
# 버블 정렬은 인접한 두 값을 비교해서 순서가 틀리면 서로 교환
# (swap) 하는 정렬입니다.
```

```python
arr = [5, 2, 8, 1]

for i in range(len(arr)):
    for j in range(len(arr)-1):
        if arr[j] > arr[j+1]:
            temp = arr[j]
            arr[j] = arr[j+1]
            arr[j+1] = temp

print(arr)
```
