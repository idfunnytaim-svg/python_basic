```python
a = 10
print(a)
print(type(a))#함수를 호출하는 연산자 type뒤에 ()를쓰는게 호출하는거다.
```

```python
# 2. 부동소수점수 (float): 오차를 가진다.
b = 3.14
print(b)
print(type(b))
print(0.1+0.2)
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
print(3>5)
```

```python
# 5. 문자열 (str)-자체가 객체여서 변경이 안되어서 타입에어가 뜬다.
e = "Python"
print(e)
print(type(e))
print(e[1])
e[1]='k'
```

```python
# 6. 리스트 (list)- 객체 여러개 가지고 있는애
f = [1, 2, 3]
print(f)
print(type(f))
```

```python
# 7. 튜플 (tuple)
g = (1, 2, 3)
print(g)
print(type(g))
```

```python
g[1]=9#읽기전용 리스트
```

```python
# 8. 딕셔너리 (dict)
h = {"a": 1}
print(h)
print(type(h))
```

```python
# 9. 집합 (set)-인덱스도 없고, 키도 없다
i = {1, 2, 3}
print(i)
print(type(i))
```

```python
print(i[0])
```

```python
# 10. 범위 (range)
j = range(5)
print(j)
print(type(j))
print(list(j))
```
